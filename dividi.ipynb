## Divido il dataset in due file: uno che contiene i canali e uno che contiene i nodi

```python
import json
import pandas as pd
import networkx as nx
```

```python
new_filename_n = 'nodes.json'
```

```python
with open ('dataset_ln_del_01-05-21.json','r', encoding='utf-8') as data_file:
    data = json.load(data_file)
df_nodes = pd.json_normalize(data, record_path=["nodes"])
#tengo solo le prime 5 colonne
df_drop_n = df_nodes.drop(df_nodes.iloc[:,5:68],axis=1)

#df_channels = pd.json_normalize(data_channels, record_path=["nodes","addresses"], meta=[["nodes","alias"],["nodes","pub_key"]])
```

```python
display(df_drop_n)
```

```python
to_write_n = df_drop_n.to_json(orient='records', indent = 1)
with open (new_filename_n, "w") as file:
    file.write(to_write_n)
```

```python
new_filename_e = 'edges.json'
```

```python
df_edges = pd.json_normalize(data, record_path = ["edges"])
```

```python
display(df_edges)
```

```python
to_write_e = df_edges.to_json(orient='records', indent = 1)
with open(new_filename_e,"w") as file1:
    file1.write(to_write_e)
```
